```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
duration_s = 0.02
sample_rate_Hz = 192000
tone1_Hz = 700
tone2_Hz = 1900

Nsamples = duration_s * sample_rate_Hz
t = np.arange(0,Nsamples)/sample_rate_Hz
y1 = 0.5*np.sin(2*np.pi*tone1_Hz*t)
y2 = 0.5*np.sin(2*np.pi*tone2_Hz*t)
ys = y1+y2

z1 = 0.5*np.cos(2*np.pi*tone1_Hz*t)
z2 = 0.5*np.cos(2*np.pi*tone2_Hz*t)
zs = z1+z2

np.savetxt("/home/oliver/Sync/Ham/T41/Software/T41-V12-SDT/testing/two_tone_700-1900_Hz_192000ksps_0.02s-I.txt",ys)
np.savetxt("/home/oliver/Sync/Ham/T41/Software/T41-V12-SDT/testing/two_tone_700-1900_Hz_192000ksps_0.02s-Q.txt",zs)
```

```python
plt.figure()
plt.plot(t,y1)
plt.plot(t,y2)
plt.xlim([0,0.01])
plt.figure()
plt.plot(t,ys)
plt.plot(t,zs)
#plt.xlim([0,0.05])
```

```python
CW_power = np.array([[0,+3.9],
[10,-5.7],
[20,-15.4],
[30,-25.3]])
plt.figure(facecolor='white')
plt.plot(CW_power[:,0],CW_power[:,1])
plt.grid()
plt.xlabel("Attenuation [dB]")
plt.ylabel("Output power [dBm]")
plt.title("RF board output power in CW mode")
```

```python
def fit_line(x,y):
    A = np.vstack([x, np.ones(len(x))]).T
    m, c = np.linalg.lstsq(A, y)[0]
    return m,c
x = CW_power[:,0]
y = CW_power[:,1]
m,c = fit_line(x,y)
```

```python
plt.figure(facecolor='white')
plt.plot(CW_power[:,0],CW_power[:,1],'ko',label='data')
plt.plot(CW_power[:,0],m*CW_power[:,0]+c,'k--',label="%3.2f*Att + %3.2f"%(m,c))
plt.legend()
plt.grid()
plt.xlabel("Attenuation [dB]")
plt.ylabel("Output power [dBm]")
plt.title("RF board output power in CW mode")
plt.savefig("RFboard_power_out_in_CW_mode.png")
```

```python
a = """Microphone amplitude [mV]	IQ amplitude [mV]	Gain
3	90.5	30.2
5	160	32.0
10	330	33.0
20	673	33.7
30	1025	34.2
"""
dat = [b.split("\t") for b in a.split("\n")][1:-1]
dat = np.array(dat).astype(float)

def fit_line(x,y):
    A = np.vstack([x, np.ones(len(x))]).T
    m, c = np.linalg.lstsq(A, y)[0]
    return m,c
x = dat[:,0]
y = dat[:,1]
m,c = fit_line(x,y)

plt.figure(facecolor='white')
plt.plot(dat[:,0],dat[:,1],'ko',label='data')
plt.plot(dat[:,0],m*dat[:,0]+c,'k--',label="%3.2f*input %3.2f"%(m,c))
plt.legend()
plt.grid()
plt.xlabel("Microphone input [mV]")
plt.ylabel("IQ amplitude [mV]")
plt.title("Gain of the main board DSP chain")
plt.savefig("Mainboard_IQ_out_vs_mic_in.png")
```
